```python
# Superstore Sales Data Analysis
#import
import pandas as pd
import matplotlib.pyplot as plt

#Load Data
df = pd.read_csv(r"D:\Ashora Francis\project\datasets\superstore sale\train.csv")
df.head()

#Data Understanding
df.info()
df.describe()

#Data Cleaning
df.isnull().sum()
df = df.dropna()
df.isnull().sum()

#Analysis
df['Category'].value_counts()
df.groupby('Category')['Sales'].sum()
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
df.groupby('Region')['Sales'].sum()

#Visualization sales by region
plt.figure(figsize=(6,4))
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()

#sales by category 
plt.figure(figsize=(6,4))
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()

#Top 5 Products
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
#Sales by Segment
df.groupby('Segment')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Segment")
plt.show()
```

```python
1. Sales are highest in the West and lowest in the South.

2. The category with the highest sales is technology, which is followed by office supplies and furniture.

3. The Home Office segment contributes the least to overall sales, whereas the Consumer segment contributes the most.

4. Customer segment and region have a big impact on sales.

5. To boost sales, the company ought to concentrate more on the South.
```
